```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
y
```

```python
# functional method

plt.plot(x,y,"r")
plt.xlabel("X Lable")
plt.ylabel("Y Lable")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
# object-oriented method

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X Lable")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.15,0.4,0.3])
# 시작 x좌표,y좌표,x축길이,y축길이 
axes1.plot(x,y)
axes1.set_title("Bigger plot")
axes2.plot(y,x)
axes2.set_title("Smaller plot")
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

# for current_ax in axes:
#     current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title("First title")

axes[1].plot(y,x)
axes[1].set_title("Second title")

plt.tight_layout()
```

```python
# figure size and DPI

# fig = plt.figure(figsize=(8,2))
fig,axes = plt.subplots(nrows =2,ncols =1,figsize=(8,2))

# ax = fig.add_axes([0,0,1,1])
# ax.plot(x,y)
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig("my_picture.png",dpi=200)
```

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X Squred")
ax.plot(y,x**3, label = "X Cubed")

ax.legend(loc=0)
```
